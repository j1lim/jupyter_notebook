```python
combined_dir = "/media/sf_j1/MAP/RNA-seq/analysis/ortholog_diff_strain/"
temp_files = os.listdir(combined_dir)
mouse_infection_files = []
for x in temp_files:
    if "ATCC.tsv" in x:
        mouse_infection_files.append(x)
mouse_infection_files.sort()
print mouse_infection_files

for x in mouse_infection_files:
    sample_dir = combined_dir + x
    df_exp = pd.read_csv(sample_dir, sep = "\t")

    fig = plt.figure( figsize=(7, 7) )

    lst_changed_atcc = df_exp.FPKM_media
    lst_original_atcc = df_exp.FPKM_mouse_infection
    lst_color = abs( ( lst_changed_atcc - lst_original_atcc ) / ( lst_changed_atcc + lst_original_atcc ) )

    ax = plt.subplot()    
    ax.set_title( "ATCC mouse infection vs.media" )
    ax.set_yscale( "log" )
    ax.set_xscale( "log" )

    ax.scatter( lst_changed_atcc, lst_original_atcc, s = 20, alpha = 0.8, c = lst_color )
    ax.set_ylabel( "mouse_infection" )
    ax.set_xlabel( "media" )
    ax.set_xlim( [ 0.01, 100000 ] )
    ax.set_ylim( [ 0.01, 100000 ] )

    plt.draw()
    figure_name = x.replace("combined_", "")[:-4]
    plt.savefig( "/media/sf_j1/MAP/RNA-seq/figure/%s.svg" % figure_name )
    plt.savefig( "/media/sf_j1/MAP/RNA-seq/figure/%s.png" % figure_name )
    plt.show()
```
