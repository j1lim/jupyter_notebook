```python
import os
import primer3
from Bio import AlignIO
dir_notebook = os.getenv( "HOME" )
file_include = dir_notebook + "/_include.ipynb"
%run $file_include
```

```python
temp_name = "Org112_Gene894"
file_alignment = "/media/sf_j1/MAP/analyzed_data/MAP/MAP_40/only_MAP/temp/%s.clustal_num" % temp_name
alignment = AlignIO.read( open( file_alignment, "r" ), "clustal" )

alignment_len = alignment.get_alignment_length()
alignment_count = len( alignment )

print "alignment_count=%d" % alignment_count, "alignment_length=%d" % alignment_len
```

```python
count_threshold = alignment_count * 1.0
dic_ntd = {}
seq_motif = ""

for i in range( alignment_len ) :
    dic_ntd0 = { "A": 0, "T": 0, "G": 0, "C": 0, "-": 0 }
    for j in range( alignment_count ) :
        seq_align = alignment[ j ]
        seq_align = str( seq_align.seq )
        char_align = seq_align[ i ]
        if char_align in dic_ntd0.keys() :
            dic_ntd0[ char_align ] = dic_ntd0[ char_align ] + 1

    max_count = max( dic_ntd0.values() )
    if max_count >= count_threshold :
        for k, v in dic_ntd0.items() :
            if max_count == v :
                seq_motif = seq_motif + k
    else :
        seq_motif = seq_motif + "N"
    dic_ntd[ i ] = dic_ntd0    

seq_motif = seq_motif.replace("-", "N")
print seq_motif
```

```python
primer3.bindings.designPrimers(
        {
            'SEQUENCE_ID': '...',
            'SEQUENCE_TEMPLATE': seq_motif
        },
        {
            'PRIMER_OPT_SIZE': 20,
            'PRIMER_PICK_INTERNAL_OLIGO': 1,
            'PRIMER_INTERNAL_MAX_SELF_END': 8,
            'PRIMER_MIN_SIZE': 18,
            'PRIMER_MAX_SIZE': 25,
            'PRIMER_OPT_TM': 60.0,
            'PRIMER_MIN_TM': 57.0,
            'PRIMER_MAX_TM': 63.0,
            'PRIMER_MIN_GC': 20.0,
            'PRIMER_MAX_GC': 80.0,
            'PRIMER_MAX_POLY_X': 100,
            'PRIMER_INTERNAL_MAX_POLY_X': 100,
            'PRIMER_SALT_MONOVALENT': 50.0,
            'PRIMER_DNA_CONC': 50.0,
            'PRIMER_MAX_NS_ACCEPTED': 0,
            'PRIMER_MAX_SELF_ANY': 12,
            'PRIMER_MAX_SELF_END': 8,
            'PRIMER_PAIR_MAX_COMPL_ANY': 12,
            'PRIMER_PAIR_MAX_COMPL_END': 8,
            'PRIMER_PRODUCT_SIZE_RANGE': [230, 270],
        })
```

```python
f_w = open("/media/sf_j1/MAP/analyzed_data/MAP/MAP_40/only_MAP/primer_design/primer_candidate/avium_%s_primer" % temp_name, "w")

result = primer3.bindings.designPrimers(
    {
        'SEQUENCE_ID': '...',
        'SEQUENCE_TEMPLATE': seq_motif
    },
    {
        'PRIMER_OPT_SIZE': 20,
        'PRIMER_PICK_INTERNAL_OLIGO': 1,
        'PRIMER_MIN_SIZE': 18,
        'PRIMER_MAX_SIZE': 25,
        'PRIMER_OPT_TM': 60.0,
        'PRIMER_MIN_TM': 57.0,
        'PRIMER_MAX_TM': 63.0,
        'PRIMER_MIN_GC': 20.0,
        'PRIMER_MAX_GC': 80.0,
        'PRIMER_MAX_POLY_X': 100,
        'PRIMER_INTERNAL_MAX_POLY_X': 100,
        'PRIMER_SALT_MONOVALENT': 50.0,
        'PRIMER_DNA_CONC': 50.0,
        'PRIMER_MAX_NS_ACCEPTED': 0,
        'PRIMER_MAX_SELF_ANY': 12,
        'PRIMER_MAX_SELF_END': 8,
        'PRIMER_PAIR_MAX_COMPL_ANY': 12,
        'PRIMER_PAIR_MAX_COMPL_END': 8,
        'PRIMER_PRODUCT_SIZE_RANGE': [230, 270],
    })

lst_keys = result.keys()
lst_keys.sort()
for y in range(5):
    y_ = str(y)
    for x in lst_keys:
        if y_ in x:
            print "{0:34s}".format(x), ":", result[x]
            f_w.write("{0:33s}".format(x) + "\t: " + str(result[x]) + "\r\n")
    print ""
    f_w.write("\r\n")
f_w.close()
```
